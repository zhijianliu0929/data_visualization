```python
import matplotlib.pyplot as plt
import numpy as np
```

### this enables you see the plot inside the Jupyter Notebook, otherwise, you need to use plotshow()

```python
%matplotlib inline 
```

```python
x = np.linspace(0,6,11)
y = x**2
```

```python
plt.plot(x,y, '-r')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)
# 1: number of rows, number of column, and which plot you are plotting
plt.plot(x,y,'r')

plt.subplot(1,2,2)
# 1: number of rows, number of column, and which plot you are plotting
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
# a figure object is created. a blank canvas is created.
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # left, buttom, right, up
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
# this method treat the figure as an object
```
