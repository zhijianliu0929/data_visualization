```python
import matplotlib.pyplot as plt
import numpy as np
```

### this enables you see the plot inside the Jupyter Notebook, otherwise, you need to use plotshow()

```python
%matplotlib inline 
```

```python
x = np.linspace(0,6,11)
y = x**2
```

```python
plt.plot(x,y, '-r')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)
# 1: number of rows, number of column, and which plot you are plotting
plt.plot(x,y,'r')

plt.subplot(1,2,2)
# 1: number of rows, number of column, and which plot you are plotting
plt.plot(y,x,'b')
```

```python
fig = plt.figure(figsize = (6,6))
# The default figure size is (6.4, 4.8) inches
# a figure object is created. a blank canvas is created.
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # left, buttom, weight, highth
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
# this method treat the figure as an object
```

```python
fig = plt.figure()
# a figure object is created. a blank canvas is created.
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # left, buttom, weight, height
axes3 = fig.add_axes([0.2,0.5,0.4,0.35]) # left, buttom, width, height
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.plot(x,y)
axes3.plot(y,x)
axes1.set_title('Larger Plot')
axes3.set_title('Smaller Plot')
# this method treat the figure as an object
```

```python
fig,axes = plt.subplots(nrows=1,ncols =2)
axes[0].plot(x,y)
axes[0].set_title('First Figure')
axes[1].plot(y,x)
axes[1].set_title('Second Figure')
plt.tight_layout() # take care of overlapping (text and figure can overlap), always at the end of plot
```

```python
fig.savefig('my_fig.png', dpi =200)
# Dots per inches (dpi) determines how many pixels the figure comprises. The default dpi in matplotlib is 100
```

```python
fig =plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'X Squared', color = '#FF8C00', linewidth =3, linestyle = "-",alpha = 0.9) # alpha = transparency
ax.plot(x,x**3, label = 'X Cubed', lw = 5,  marker = 'o', markersize=10) 

# Add a legend
ax.legend(loc = (0.1,0.1))
```
